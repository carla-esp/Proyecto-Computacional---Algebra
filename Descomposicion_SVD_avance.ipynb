```python
#Importamos libreria
import numpy as np

def truncated_svd(A, k):
    U, S, Vh = np.linalg.svd(A, full_matrices=False)
    return U[:, :k], S[:k], Vh[:k]
```

**3.** Muestre una tabla donde se compare el error cuadrático medio de las aproximaciones encontradas en los items anteriores (es decir, para los valores de k mencionados, y tanto para el caso centrado como el no centrado). Recuerde que este error esta dado por la fórmula

$ECM(A^{(k)} ,A) = \frac{|| A^{(k)} - A ||_{F}^{2}}{m}$

donde A(k) contiene las proyecciones de las imagenes como filas, y m es la cantidad de imágenes. Comente.

Realizaremos nuevamente las proyecciones realizadas en los items anteriores, sin embargo la diferencia ahora radica en que utilizaremos una lista para almacenar el Error Cuadratico Medio, realizaremos los calculos con la formula del enunciado, y almacenaremos el valor resultante en una lista. Despues, mostraremos todos estos errores en una tabla.

```python
#Listas donde guardaremos los errores calculados
error_no_centrado = []
error_centrado = []

#"m" es la cantidad de imágenes, es decir el número de filas en A
m = A.shape[0]

#CASO 1: SVD sin centrar (como en el punto 1)
for k in valores_k:
    #Calculamos la descomposición SVD
    U, S, Vh = truncated_svd(A, k)
    
    #Reconstruimos la matriz A usando solo las k componentes
    A_k = U @ np.diag(S) @ Vh
    
    #Calculamos el Error Cuadrático Medio (ECM)
    ## np.linalg.norm(..., 'fro') calcula la norma de Frobenius
    ECM = (np.linalg.norm(A - A_k, 'fro') ** 2) / m
    
    #Guardamos el resultado en la lista
    error_no_centrado.append(ECM)

 
#CASO 2: Matriz a la cual le aplicamos PCA (como en el punto 2)
#Primero centramos los datos (restamos la media de cada columna)
A_media = np.mean(A, axis=0)
A_centrado = A - A_media

for k in valores_k:
    #Descomponemos la matriz centrada con SVD
    U, S, Vh = truncated_svd(A_centrado, k)
    
    ## Reconstruimos la versión centrada
    A_k_centrado = U @ np.diag(S) @ Vh
    
    #Descentramos (volvemos a sumar la media)
    A_k = A_k + A_media
    
    ## Calculamos el error ECM nuevamente
    ECM = (np.linalg.norm(A - A_k, 'fro') ** 2) / m
    
    ## Guardamos el resultado
    error_centrado.append(ECM)

#Importamos Pandas para realizar la tabla
import pandas as pd
#Creamos una tabla comparativa con los resultados
tabla_errores = pd.DataFrame({
    'k': valores_k,
    'ECM_no_centrado': error_no_centrado,
    'ECM_centrado': error_centrado
})

# Mostramos la tabla
print(tabla_errores)
```

## Inciso (a)

```python
votes_a = votes.copy()
```

```python
# Pedimos que en cada fila numérica se obtenga el promedio, este lo redondee al primer decimal
# y reemplace este valor por los NaN
votes_a = votes_a.fillna(votes_a.mean(numeric_only=True).round(1))
```

```python
votes_a.isna().sum()
```

```python
votes_a.head()
```

```python
# La centralización de datos, permite reducir las dimension en la que se trabaja 
# Cada columna tiene una media aprox de 0
X = votes_a.select_dtypes(include=['number'])
X_centrado = X - X.mean()
```

```python
votes_a['party'].unique()
```

```python
# Formamos la matriz
Matrix = X_centrado.to_numpy()

# Hacemos la descomposición SVD
U, S, Vt = np.linalg.svd(X_centrado, full_matrices=False)

# La dos primeras filas de la matriz V (direcciones)
V2 = Vt[:2, :]

# La proyección indica donde se centran más los votos si miramos un plano
X_proj = Matrix @ V2.T
```

```python
plt.figure(figsize=(7,5))

# eje x, eje y, colores
plt.scatter(X_proj[:,0], 
            X_proj[:,1],
            c = votes_a['party'].astype('category').cat.codes)

plt.xlabel('Primera dirección SVD')
plt.ylabel('Segunda dirección SVD')
plt.title('Visualización sobre las posiciones políticas SVD')
plt.show()
```
